```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
t_max = 4  # Tiempo máximo

# Valores de Delta t
dt_values = [0.2]
colors = ['green']  # Colores para los diferentes dt

# Solución analítica
def analytical_solution(t):
    return np.exp(t)

time_analytical = np.linspace(0, t_max, 100)
solution_analytical = analytical_solution(time_analytical)

# Gráfica
plt.figure(figsize=(8, 5))
plt.plot(time_analytical, solution_analytical, label='Solución analítica', color='blue')

# Implementación del método de Euler para cada dt
for i, dt in enumerate(dt_values):
    time_steps = np.arange(0, t_max + dt, dt)
    u = np.zeros(len(time_steps))
    u[0] = 1  # Condición inicial
    
    for n in range(len(time_steps) - 1):
        u[n + 1] = u[n] + dt * u[n]  # Método de Euler explícito
    
    plt.scatter(time_steps, u, label=f'dt={dt}', color=colors[i])

# Configuración de la gráfica
plt.xlabel('Time')
plt.ylabel('u(t)')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler(dt, t):
    pasos = np.arange(0, t + dt, dt)
    
    # Inicializo mi array 
    u = np.zeros(len(pasos))
    u[0] = 1  # Condición inicial
    
    for i in range(len(pasos) - 1):  # Se corrigió el rango del bucle
        u[i + 1] = u[i] + dt * u[i]
    
    return pasos, u

# Parámetros
t = 4  # Tiempo total
dt = 0.2  # Intervalo de tiempo

pasos, u = euler(dt, t)

def solucion_analitica(t):
    return np.exp(t)

tiempo_analitico = np.linspace(0, t, 100)
solucion_analitic = solucion_analitica(tiempo_analitico)


# Graficar resultados
plt.figure(figsize=(8, 5))
plt.plot(tiempo_analitico, solucion_analitic, label='Solución analítica', color='blue')
plt.scatter(pasos, u, color='red', label='Euler dt=0.2')
plt.xlabel('Time')
plt.ylabel('u(t)')
plt.title('Taller 1')
plt.legend()
plt.grid()
plt.show()
```
